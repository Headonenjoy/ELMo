## dimensionality_reduction_by_PCA

```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
# make_blobs生成聚类模型数据
from sklearn.datasets import make_blobs
```

```python
# 关键参数有n_samples（生成样本数）， n_features（样本特征数），centers(簇中心的个数或者自定义的簇中心)和cluster_std（簇数据方差，代表簇的聚合程度）
# X为样本特征，Y为样本簇类别， 共1000个样本，每个样本4个特征，共3个簇，簇中心在[-1, -1, -1], [1, 1, 1], [2, 2, 2]， 簇方差分别为[0.4, 0.5, 0.2]
X_3d, y = make_blobs(n_samples=1000, n_features=3, centers=[[-1, -1, -1], [1, 1, 1], [2, 2, 2]], cluster_std=[0.4, 0.5, 0.2])
```

```python
#绘制原始数据的三维点状图
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_3d[:, 0], X_3d[:, 1], X_3d[:, 2],c=y, s=3)
plt.show()
```

```python
#绘制三维数据特征前两维度
plt.scatter(X_3d[:, 0], X_3d[:, 1], c=y, s=3, marker='o')
plt.show()
```

```python
#绘制三维数据特征前一维度
plt.scatter(range(len(X_3d)), X_3d[:, 1], c=y, s=3, marker='o')
plt.show()
```

```python
def change_color(y):
    y2 = list()
    for item in y:
        if item == 0:
            y2.append(2)
        elif item == 1:
            y2.append(1)
        else:
            y2.append(0)
    return y2

y2 = change_color(y)
```

```python
#绘制经过PCA降维后的二维数据
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_3d)
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=y2, s=3, marker='o')
print(" pca.n_components: ", pca.n_components)
print(" pca.explained_variance_: ", pca.explained_variance_)
print(" pca.explained_variance_ratio_: ", pca.explained_variance_ratio_)
plt.show()
```

```python
#绘制经过PCA降维后的一维数据
pca = PCA(n_components=1)
X_2d = pca.fit_transform(X_3d)
plt.scatter(list(range(len(X_2d))), X_2d, c=y2, s=3, marker='o')
print(" pca.n_components: ", pca.n_components)
print(" pca.explained_variance_: ", pca.explained_variance_)
print(" pca.explained_variance_ratio_: ", pca.explained_variance_ratio_)
plt.show()
```
